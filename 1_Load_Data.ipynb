## Filter Subset: train

```python
import requests
import json
import csv

# URL of the JSON file
url = "https://datarepo.eng.ucsd.edu/mcauley_group/gdrive/googlelocal_restaurants/filter_all_t.json"

# Send a GET request to the URL to fetch the JSON data
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON data
    json_data = json.loads(response.text)['train']
    
    # Extract the keys (assuming JSON data is in the form of a list of dictionaries)
    keys = json_data[0].keys()
    
    # Path to save the CSV file
    csv_file_path = "data_train.csv"
    
    # Write the JSON data to a CSV file
    with open(csv_file_path, 'w', newline='') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=keys)
        
        # Write the header
        writer.writeheader()
        
        # Write the rows
        for data in json_data:
            writer.writerow(data)
    
    print("CSV file saved successfully!")
else:
    # If the request was unsuccessful, print an error message
    print("Failed to fetch data from the URL:", response.status_code)
```

## Image Review All

```python
import requests
import json
import csv

# URL of the JSON file
url = "https://datarepo.eng.ucsd.edu/mcauley_group/gdrive/googlelocal_restaurants/image_review_all.json"

# Send a GET request to the URL to fetch the JSON data
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Split the response text into individual JSON objects or arrays
    json_data_list = response.text.split('\n')  # Change '\n' to appropriate delimiter if needed

    # Extract the keys (assuming JSON data is in the form of a list of dictionaries)
    keys = json.loads(json_data_list[0]).keys()
    
    # Path to save the CSV file
    csv_file_path = "data_all.csv"
    
    # Write the JSON data to a CSV file
    with open(csv_file_path, 'w', newline='') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=keys)
        
        # Write the header
        writer.writeheader()
        
        # Write the rows
        for json_data in json_data_list:
            if json_data.strip():  # Check if JSON data is not empty
                # Parse the JSON data
                data_dict = json.loads(json_data)
                
                # Write the row
                writer.writerow(data_dict)
    
    print("CSV file saved successfully!")
else:
    # If the request was unsuccessful, print an error message
    print("Failed to fetch data from the URL:", response.status_code)
```
